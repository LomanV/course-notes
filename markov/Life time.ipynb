```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

"""
This walk illustrates recurrency of Markov chains. This chain represents the lifetime of component,
a(i, aplha) represents the probability that it lives from time i to time i+1 depending on alpha,
otherwise it dies and his replaced at time i+1 by a new component of age 0. 
Here the chain is reccurent if and only if alpha <= 1
"""
```

```python
T = 10000 # Length of a walk
K = 10 # Number of walks
alpha_vec = [0.8, 1., 1.01, 1.2] 

fig, axs = plt.subplots(2, 2)
fig.set_size_inches(18.5, 10.5)
l, j = 0, 0
for alpha in alpha_vec:
    x = np.linspace(0,T-1, T)
    for k in range(K):
        walk = np.zeros(T)
        for i in range(T-1):
            walk[i+1] = (walk[i]+1)*(np.random.rand()>1./((walk[i]+2)**alpha))
        axs[l, j].plot(x, walk)
        axs[l, j].set_title("K = " +str(K) + " walks of length T = " + str(T) + " with alpha = " + str(alpha))
    if j == 0:
        j = j + 1
    else:
        j, l = 0, 1
plt.show()
```
