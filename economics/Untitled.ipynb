```python
import numpy as np
import math
```

```python
A0 = 1
gL = .3
gA = .3
alpha = .5
beta = 1 - alpha
s = .2
delta = .1

# First constant approximation
A1 = 5
L1 = 6
```

```python
def K(t):
    return t
K = np.vectorize(K)

def L(t):
    return A0*math.exp(gL*t)
L = np.vectorize(L)

def A(t):
    return A0*math.exp(gA*t)
A = np.vectorize(A)

def F(K,L,A):
    return (K**alpha)*(A*L)**beta
F = np.vectorize(F)

t = np.linspace(0,10,50)
```
